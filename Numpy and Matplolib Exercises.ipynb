```python
# Making Two Sinusoid Functions
# I just plotted as asked and search for "how to make plot 
# have read colors" and the like on Google.


import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1,20,0.1)
a = np.sin(2*x+1)
b = np.sin(4*x+6)
c = a*b

plt.plot(x, c, label='a * b')
plt.plot(x, a, label='a = sin(2x+1)', alpha=0.3)
plt.plot(x, b, label='b = sin(4x+6)', alpha=0.3)
plt.title('Sample Numpy and Matplot Lib Exercises')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.legend()
plt.show()
```

```python
# Points within a Circle

# I first plotted the circle, and generated coordinates.
# Out of those coordinates, I made a list on ones that fall inside and outside the
# quarter circle using the condition of the equation of the circle.
# Then I just counted the lengths of the list to know how many fell within and outside the quarter circle.

import numpy as np
import matplotlib.pyplot as plt 

x = np.linspace(0,1,100) # 100 for a smoother curve
y = np.sqrt(1**2 - x**2)

plots = 10000
a = np.random.uniform(0,1,plots)
b = np.random.uniform(0,1,plots)

inside_x=[]
inside_y=[]
outside_x=[]
outside_y=[]

for i in range(plots):
    if a[i]**2 + b[i]**2 <= 1:
        inside_x.append(a[i])
        inside_y.append(b[i])
    else:
        outside_x.append(a[i])
        outside_y.append(b[i])

plt.scatter(inside_x,inside_y,color='yellowgreen',s=1,label='Plots inside the QC')
plt.scatter(outside_x,outside_y,color='lightgray',s=1,label='Plots outside the QC')
plt.plot(x,y, label='Radius 1',color='red',linestyle='--')
plt.gca().set_aspect('equal')
plt.title('Points within a Quarter Circle')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(True)
plt.legend()
plt.show()

print(f'Total plots: {plots}')
print(f'Plots inside: {len(inside_x)}')
print(f'Plots outside: {len(outside_x)}')
```

```python
# Guassian
# I just generated numbers from mean 0 and stdev 1
# then plotted them on a histogram with 50 bars

quantity = 2000

data = []
for i in range(quantity):
    data.append(np.random.normal(0,1))

normalized = []
for j in range(quantity):
    normalized.append((data[j]-min(data))/(max(data)-min(data)))

plt.figure(1)
plt.hist(normalized,bins=50,color='pink',edgecolor='hotpink')
plt.title('Histogram of Drawn Numbers from N(0,1)')
plt.xlabel('Normalized Numbers')
plt.ylabel('Frequency of Occurence')

plt.figure(2)
plt.hist(data,bins=50,color='skyblue',edgecolor='royalblue')
plt.title('Histogram of Drawn Numbers from N(0,1)')
plt.xlabel('Raw Numbers')
plt.ylabel('Frequency of Occurence')

plt.show()
```
