```python
import matplotlib.pyplot as plt

def collatz(N):
    length = 0
    while N != 1:
        if N % 2 == 0:
            N = N / 2
        elif N % 2 == 1: 
            N = 3*N + 1
        length += 1
    else:
       return length

N = range(1, 10000)
T = [collatz(n) for n in N]
     
plt.scatter(N, T, 5)
plt.xlabel("N")
plt.ylabel("T(N)")
plt.title ("Collatz Sequence T(N) vs. N")
plt.show()
```
